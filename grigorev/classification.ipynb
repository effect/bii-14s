```python
import numpy, scipy
from sklearn.utils import shuffle
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plot
%matplotlib inline
```

```python
# load data
blue = numpy.loadtxt("./data/blue.txt")
red = numpy.loadtxt("./data/red.txt")
```

```python
# visualize data
_ = plot.plot(blue[:, 0], blue[:, 1], "b.")
_ = plot.plot(red[:, 0], red[:, 1], "r.")
```

```python
# add classes to the matrix
BLUE = 0
RED = 1
full_set = numpy.concatenate((
        numpy.hstack((
            blue,
            [[BLUE]] * len(blue)
        )),
        numpy.hstack((
            red,
            [[RED]] * len(red)
        ))
    ),
    axis = 0
)
```

```python
# shuffle data to avoid possible batch effects and whatnot
X = full_set[:, :2] # coordinates
Y = full_set[:, 2] # classes
X, Y = shuffle(X, Y, random_state = 1)
```

```python
# split data into a training set and a testing set
TRAIN_FRACTION = 0.6
TRAIN_SIZE = int(full_set.shape[0] * TRAIN_FRACTION)
train_X = X[:TRAIN_SIZE]
train_Y = Y[:TRAIN_SIZE]
test_X = X[TRAIN_SIZE:]
test_Y = Y[TRAIN_SIZE:]
```

```python
# perform a K-nearest Neighbors classification
KNN = KNeighborsClassifier(n_neighbors = 5)
KNN.fit(train_X, train_Y)
KNN_score = KNN.score(train_X, train_Y)
KNN_test_score = KNN.score(test_X, test_Y)
print('Training set score = ' + str(KNN_score))
print('Test set score = ' + str(KNN_test_score))
```

```python
# visualize misses of KNN
KNN_misses = []
for X, Y in zip(test_X.tolist(), test_Y.tolist()):
    if KNN.predict(X) != Y:
        KNN_misses.append(X)
KNN_misses = numpy.array(KNN_misses)
print("KNN wrongly classified " + str(len(KNN_misses)) + " test points (shown in red):")
_ = plot.scatter(test_X[:, 0], test_X[:, 1], color = "black")
_ = plot.scatter(KNN_misses[:, 0], KNN_misses[:, 1], color = "red")
```

```python
# perform a Random Forest classification
RF = RandomForestClassifier(n_estimators = 10)
RF.fit(train_X, train_Y, sample_weight = None)
RF_score = RF.score(train_X, train_Y)
RF_test_score = RF.score(test_X, test_Y)
print('Training set score = ' + str(RF_score))
print('Test set score = ' + str(RF_test_score))
```

```python
# visualize misses of RF
RF_misses = []
for X, Y in zip(test_X.tolist(), test_Y.tolist()):
    if RF.predict(X) != Y:
        RF_misses.append(X)
RF_misses = numpy.array(RF_misses)
print("RF wrongly classified " + str(len(RF_misses)) + " test points (shown in red):")
_ = plot.scatter(test_X[:, 0], test_X[:, 1], color = "black")
_ = plot.scatter(RF_misses[:, 0], RF_misses[:, 1], color = "red")
```

<b><u>Заключение:</u></b><br><br>
KNN неверно классифицировал 51 точку из тестового сета, в то время как количество ошибок RF -- 63.<br>
Также, score для тестового сета при использовании KNN больше, чем для RF (.8725 > .8425)<br>
Из этого можно сделать вывод, что, по крайней мере, в данных условиях лучшим классификатором является метод <b>Random Forest.</b>
