```python
import numpy
from sklearn.utils import shuffle
from scipy.stats import linregress
import matplotlib.pyplot as plot
%matplotlib inline
```

```python
SET_SIZE = 1000
TRAIN_SET_FRACTION = 0.8
TRAIN_SET_SIZE = int(SET_SIZE * TRAIN_SET_FRACTION)
TEST_SET_SIZE = SET_SIZE - TRAIN_SET_SIZE
```

Generate and display a random model:

```python
model_betas = numpy.random.random_sample(2) * 4 - 2
"Original model: y = {0} + {1}*x".format(model_betas[0], model_betas[1])
```

Generate 1000 points corresponding to the model and blur them by introducing normally distributed errors:

```python
# points according to model
X = numpy.linspace(0, 1, SET_SIZE)
y = model_betas[0] + model_betas[1] * X
# draw model
_ = plot.plot(X, y, color = "blue", label = "Original model")
# introduce error
for index, _ in enumerate(y):
    y[index] += numpy.random.randn()
# draw points
_ = plot.plot(X, y, "g.", label = "Points with errors")
# display label
_ = plot.legend(loc = "best")
```

Divide the set into a training set and a testing set:

```python
# make data unordered
X, y = shuffle(X, y, random_state = 1)
# pick training set
train = lambda: None
train.X = X[:TRAIN_SET_SIZE]
train.y = y[:TRAIN_SET_SIZE]
# visualize training set
_ = plot.plot(
    train.X,
    train.y,
    "g.",
    label = "Training set"
)
# pick testing set
test = lambda: None
test.X = X[TRAIN_SET_SIZE:]
test.y = y[TRAIN_SET_SIZE:]
# visualize testing set
_ = plot.plot(
    test.X,
    test.y,
    "yo",
    label = "Testing set"
)
# display legend
_ = plot.legend(
    bbox_to_anchor = (1.05, 1),
    loc = 2,
    borderaxespad = 0
)
```

Perform linear regression on the training set and calculate RSS/n for both sets:

```python
# perform linear regression
betas, residuals, _, _, _ = numpy.polyfit(
    train.X,
    train.y,
    deg = 1,
    full = True
)
betas = betas[::-1]
# proposed model as function
f = lambda x: betas[0] + betas[1] * x
# human-readable result
print("Regression: y = {0} + {1}*x".format(betas[0], betas[1]))
# residual sum of squares for training set is given by polyfit
print("RSS/n for training set: {0}".format(residuals[0] / TRAIN_SET_SIZE))
# calculate residual sum of squares for testing set
testing_RSS = sum([
    (f(tx) - ty)**2
    for tx, ty in
    numpy.vstack((test.X, test.y)).T.tolist()
])
print("RSS/n for testing set:  {0}".format(testing_RSS / TEST_SET_SIZE))
```

Calculate R squared and p-value for the entire dataset:

```python
_, _, r_value, p_value, std_err = linregress(X, y)
print("R^2 = {0}\np-value = {1}".format(r_value**2, p_value))
```

Visualize training and testing datasets and original and proposed models:

```python
# draw original model
_ = plot.plot(
    train.X,
    model_betas[0] + model_betas[1] * train.X,
    color = "blue",
    label = "Original model"
)
# draw training set
_ = plot.plot(
    train.X,
    train.y,
    "g.",
    label = "Training set"
)
# draw testing set
_ = plot.plot(
    test.X,
    test.y,
    "y.",
    label = "Testing set"
)
# draw model proposed by polyfit
_ = plot.plot(
    train.X,
    f(train.X),
    color = "green",
    label = "Proposed model"
)
# display legend
_ = plot.legend(
    bbox_to_anchor = (1.05, 1),
    loc = 2,
    borderaxespad = 0
)
```

Determine which points are "good" (closer to the model than std_err), and which points are "bad" (the rest):

```python
# sort points into good and bad
good, bad = [], []
for tx, ty in numpy.vstack((X, y)).T.tolist():
    if (f(tx) - ty)**2 < std_err:
        good.append([tx, ty])
    else:
        bad.append([tx, ty])
good = numpy.array(good)
bad = numpy.array(bad)
```

Visualize. Good are green, bad are red:

```python
# draw proposed model
_ = plot.plot(
    X,
    f(X),
    color = "green",
    label = "Proposed model"
)
# draw good points
_ = plot.plot(
    good[:, 0],
    good[:, 1],
    "g.",
    label = "Good points"
)
# draw bad points
_ = plot.plot(
    bad[:, 0],
    bad[:, 1],
    "r.",
    label = "Bad points"
)
# display legend
_ = plot.legend(
    bbox_to_anchor = (1.05, 1),
    loc = 2,
    borderaxespad = 0
)
```
