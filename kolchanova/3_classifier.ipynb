Import dataset

```python
import numpy as np
red = np.loadtxt("math/red.txt")
blue = np.loadtxt("math/blue.txt")
print red.shape
print blue.shape
```

Make plot

```python
import matplotlib.pyplot as pl
pl.prism()
pl.scatter(red[:, 0], red[:, 1], c='red')
pl.scatter(blue[:, 0], blue[:, 1], c='blue')
```

Prepare data for analysis

```python
reds = np.hstack ((red, [[1]] * len (red) ))
blues = np.hstack  ((blue, [[0]] * len (blue) ))
dots = np.concatenate((reds, blues), axis=0)
x = dots[:, :-1]
y = dots[:, 2]
```

Train and test sets

```python
from sklearn.utils import shuffle
x, y = shuffle(x, y, random_state=1)
size=dots.shape[0] *0.8
x_train = x[:size]
y_train = y[:size]
x_test = x[size:]
y_test = y[size:]
```

Build clussifier

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train, y_train)
print 'Accuracy of KNN train set:', knn.score(x_train, y_train)
print 'Accuracy of KNN test set:', knn.score(x_test, y_test)
```

Predict color

```python
def predict(a,b):
    if knn.predict([a, b])==0:
        return 'blue'
    else:
        return 'red'
```
